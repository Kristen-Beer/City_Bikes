```python
# List of file names to be combined 
data_2022 = ['Resources\JC-202201-citibike-tripdata.csv', 'Resources\JC-202202-citibike-tripdata.csv', 'Resources\JC-202203-citibike-tripdata.csv', 'Resources\JC-202204-citibike-tripdata.csv', 'Resources\JC-202205-citibike-tripdata.csv', 'Resources\JC-202206-citibike-tripdata.csv', 'Resources\JC-202207-citbike-tripdata.csv', 'Resources\JC-202208-citibike-tripdata.csv', 'Resources\JC-202209-citibike-tripdata.csv', 'Resources\JC-202210-citibike-tripdata.csv', 'Resources\JC-202211-citibike-tripdata.csv', 'Resources\JC-202212-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df22_list = []

# Loop through each file in the list
for file in data_2022:
    # Read the file into a dataframe
    df22 = pd.read_csv(file)

    # Append the dataframe to the list
    df22_list.append(df22)

# Concatenate all dataframes in the list into one dataframe
df22 = pd.concat(df22_list)

# Show dataframe
df22
```
